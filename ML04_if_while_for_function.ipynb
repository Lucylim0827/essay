**문제3) 정수를 한 개 입력받아, 1부터 입력받은 수까지 각각에 대해 제곱을 구해 프린트하는 프로그램을 while loop를 사용헤 작성해 보세요. 다 작성했으면 for loop으로도 작성해봅시다.**

```python
### 실행 예시 ###
```

```python
#while loop
num = int(input("숫자를 입력하세요."))
i = 1
while i <= num:
    print(i, i*i)
    i+=1
```

```python
#for loop
num = int(input("숫자를 입력하세요."))
for i in range(1, num+1):
    print(i, i*i)
```

**문제10) 다음과 같은 긴 염기서열에서 TACCCTATGAGCTAAAAA라는 염기서열이 몇 개 있는지 찾는 함수를 for loop을 이용해 만들어 보세요.**

```python
dna_seq = 'ATCGCTAGCTAGCGATACCCTATGAGCTAAAAATTACGGCTAGCTAGTACCCTATGAGCTAAAAACTAGCTAGCTATAGCGATTACGCTAGCTAGC'
```

```python
dna_seq = 'ATCGCTAGCTAGCGATACCCTATGAGCTAAAAATTACGGCTAGCTAGTACCCTATGAGCTAAAAACTAGCTAGCTATAGCGATTACGCTAGCTAGC'
tar_seq = 'TACCCTATGAGCTAAAAA'
def count_target(dna_seq, tar_seq):
    count = 0
    dna_length = len(dna_seq)
    tar_length = len(tar_seq)
    for i in range(dna_length-tar_length+1):
        if dna_seq[i:i+tar_length] == tar_seq:
            count+=1
    return count
result = count_target(dna_seq, tar_seq)
result
```
