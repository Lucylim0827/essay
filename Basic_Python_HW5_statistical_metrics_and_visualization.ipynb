**문제1) 우리는 지난 시간에 setosa의 sepal length에 대해서 하나씩 그래프를 그렸습니다. 다른 iris type의 다른 feature에 대해 똑같이 그래프를 그려볼까요?**

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
feature = iris['data']
feature_names = iris['feature_names']
labels = iris['target']
```

```python
import pandas as pd
df = pd.DataFrame(feature, columns=feature_names)
df['target'] = labels
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
df1.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df1['petal length (cm)']
```

```python
plt.figure(figsize=(6,4))
plt.hist(df1['petal length (cm)'])
plt.title('Petal length histogram of iris versicolor', fontsize=14)
plt.show
```

```python
plt.figure(figsize=(6,4))
plt.plot(df1['petal length (cm)'])
plt.title('Petal length line graph of iris versicolor', fontsize=14)
plt.show
```

```python
import numpy as np
x = np.arange(len(df1))
```

```python
plt.figure(figsize=(6,4))
plt.bar(x, df1['petal length (cm)'])
plt.title('Petal length bar graph of iris versicolor', fontsize=14)
plt.show
```

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
ax.boxplot(df1['petal length (cm)'])
plt.title('Petal length boxplot of iris versicolor', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,4))

ratio = [len(df1['sepal width (cm)']), len(df1['petal length (cm)']), len(df1['petal width (cm)'])]
labels = ['sepal width', 'petal length', 'petal width']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.title('Pie chart of different features of iris versicolor', fontsize=14)
plt.legend(bbox_to_anchor=(1.1, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df1, x='petal length (cm)', y='petal width (cm)', hue=df1['target'], palette = 'muted')
plt.title('Scatterplot of petal length(x) and width(y) of iris versicolor', fontsize=14)
plt.show()
```

**문제2) 우리는 지난 시간에 모든 iris의 sepal length에 대해서 그래프를 겹쳐 그렸습니다. 다른 feature에 대해 똑같이 그래프를 그려볼까요?**

```python
plt.figure(figsize=(6,4))
plt.hist(df0['petal length (cm)'], alpha=0.5, label='setosa')
plt.hist(df1['petal length (cm)'], alpha=0.5, label='versicolor')
plt.hist(df2['petal length (cm)'], alpha=0.5, label='verginica')
plt.title('Petal length histrogram of all three irides', fontsize=14)
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df0['petal width (cm)'], alpha=0.5, label='setosa')
plt.hist(df1['petal width (cm)'], alpha=0.5, label='versicolor')
plt.hist(df2['petal width (cm)'], alpha=0.5, label='verginica')
plt.title('Petal width histrogram of all three irides', fontsize=14)
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
```

**문제3) 새로운 와인 데이터를 다뤄봅시다. feature 중 한 가지를 골라서 세 가지의 class를 동시에 그리는 pie chart, box plot, histogram을 그려봅시다. 우리가 그렸던 그래프와 같이 legend도 포함해주세요.**




```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine
```

```python
feature = wine['data']
feature_names = wine['feature_names']
labels = wine['target']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(feature, columns=feature_names)
df['target'] = labels
df0 = df[df['target']==0]
df1 = df[df['target']==1]
df2 = df[df['target']==2]
```

```python
plt.figure(figsize=(6,4))

ratio = [len(df0['total_phenols']), len(df1['total_phenols']), len(df2['total_phenols'])]
labels = ['class 0', 'class 1', 'class 2']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.title('Pie chart of total phenols of three classes of wine', fontsize=14)
plt.legend(bbox_to_anchor=(1.3, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
ax.boxplot([df0['total_phenols'], df1['total_phenols'], df2['total_phenols']])
plt.title('Total phenol boxplot of three classes of wine', fontsize=14)
plt.xticks(range(1,4), ['class 0', 'class 1', 'class 2'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df0['total_phenols'], alpha=0.5, label='class 0')
plt.hist(df1['total_phenols'], alpha=0.5, label='class 1')
plt.hist(df2['total_phenols'], alpha=0.5, label='class 2')
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
```
