```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['Alley', 'MasVnrType', 'MiscFeature', 'PoolQC', 'Fence', 'FireplaceQu'], axis=1, inplace=True)
```

```python
df
```

```python
set(list(df['MSZoning']))
```

```python
MSZoning_mapping = {'C (all)':0, 'FV':1, 'RH':2, 'RL':3, 'RM':4}
df['MSZoning'] = df['MSZoning'].map(MSZoning_mapping)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype == 'object':
        label_encoder = LabelEncoder()
        df[column] = label_encoder.fit_transform(df[column])
```

```python
df
```

```python
df.info()
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].mode()[0])
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].mode()[0])
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
```

```python
df
```

```python
df.drop(['Id'], axis=1, inplace=True)
```

```python
df
```

```python
X = df.iloc[:, :].values
y= df.iloc[:, 73].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
param_grid = {
    'n_neighbors' : [3, 5, 7, 9],
    'weights' : ['uniform', 'distance'],
    'p' : [1, 2],
    'metric' : ['euclidean', 'manhattan']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
rf = KNeighborsClassifier()
```

```python
kfold = KFold(n_splits=5)
grid_search = GridSearchCV(rf, param_grid, cv = kfold, scoring = 'accuracy')
```

```python
df.values
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
rf_best = KNeighborsClassifier(metric = 'manhattan',
                              n_neighbors = 7,
                              p = 1,
                              weights = 'distance')
```

```python
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_rf)
```

```python
rf_best.fit(X_train, y_train)
y_pred_rf_best = rf_best.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_rf_best)
```
